```python



# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
####### QUESTION FOR TA #### - Check with TA if this is the best way to do it
####### QUESTION FOR TA - how do you harcode NULLS


#def distinct_count(values)

unique_items = purchase_data["Item ID"].nunique()   
average_purchase_price = round(purchase_data["Price"].mean(), 2) 
total_purchases = purchase_data["SN"].count()
total_revenue = purchase_data["Price"].sum()

purchase_analysis_df = pd.DataFrame({"Unique Items": [unique_items],
                                      "Average Purchase Price": [average_purchase_price],
                                      "Total Number of Purchases": [total_purchases],
                                      "Total Revenue": [total_revenue]})
purchase_analysis_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
dedup_df = purchase_data.iloc[:,[1,3]].drop_duplicates()
gender_group_df = dedup_df.groupby("Gender")
new_df = gender_group_df.count()
new_df.rename(columns = {"SN": "Total Count"}, inplace = True)
new_df["Percentage of Players"] = round(((new_df["Total Count"]/new_df["Total Count"].sum()) * 100), 2)
new_df.sort_values(by=["Percentage of Players"], ascending = False).head()

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_group_df = purchase_data.groupby("Gender")
purchase_count = purchase_group_df["Gender"].count()

average_purchase_price = round((purchase_group_df["Price"].mean()),2)



total_purchase_value = purchase_group_df["Price"].sum()


#### NOTE - gender_group_df was built after deduping in the previous analysis
average_per_person = round((purchase_group_df["Price"].sum()/gender_group_df["Gender"].count()), 2)


purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                       "Average Purchase Price": average_purchase_price,
                                       "Total Purchase Value": total_purchase_value,
                                       "Avg Total Purchase per Person": average_per_person})


purchasing_analysis_df['Average Purchase Price'] = purchasing_analysis_df['Average Purchase Price'].map('${:,.2f}'.format)
purchasing_analysis_df['Total Purchase Value'] = purchasing_analysis_df['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_analysis_df['Avg Total Purchase per Person'] = purchasing_analysis_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)

purchasing_analysis_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#age_df = purchase_data.iloc[:,[2]]

age_dedup_df = purchase_data.iloc[:,[1,2]].drop_duplicates()
age_df = age_dedup_df.iloc[:,[1]]

age_bins = [0,9,14,19,24,29,34,39,150]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

###### QUESTION FOR TA - HOW TO TURN OFF WARNINGS ##########
age_df["age_groups"] = pd.cut(age_df["Age"], age_bins, labels=age_groups)
age_group_df = age_df.groupby("age_groups")
group_count = age_group_df.count()
group_count.rename(columns = {"Age": "Total Count"}, inplace = True)

group_count["Percentage of Players"] = round(((group_count["Total Count"]/group_count["Total Count"].sum()) * 100), 2)
group_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
price_df = purchase_data.iloc[:,[2,6]]
price_df.head()

age_bins = [0,9,14,19,24,29,34,39,150]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

price_df["age_groups"] = pd.cut(price_df["Age"], age_bins, labels=age_groups)

sl_price_df = price_df.drop(columns = ["Age"])

group_price_df = sl_price_df.groupby("age_groups")



purchase_count = group_price_df["Price"].count()
purchase_count

average_purchase_price = round(group_price_df["Price"].mean(), 2)
average_purchase_price

total_purchase_value = group_price_df["Price"].sum()
total_purchase_value

##### The below code is to calculate group counts after deduping on name
###### Defined bins again for readability

dedup_name_df = purchase_data.iloc[:,[1,2]].drop_duplicates()
dedup_name_df.head(2)

age_bins = [0,9,14,19,24,29,34,39,150]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

dedup_name_df["age_group"] = pd.cut(dedup_name_df["Age"], age_bins, labels=age_groups)
group_dedup_df = dedup_name_df.groupby("age_group")
group_dedup_df["SN"].count()

#################

average_purchase_person = round((group_price_df["Price"].sum()/group_dedup_df["SN"].count()), 2)
average_purchase_person

purchasing_age_df = pd.DataFrame({"Purchase Count": purchase_count,
                               "Average Purchase Price": average_purchase_price,
                               "Total Purchase Value": total_purchase_value,
                                 "Avg Total Purchase per Person": average_purchase_person})



purchasing_age_df['Average Purchase Price'] = purchasing_age_df['Average Purchase Price'].map('${:,.2f}'.format)
purchasing_age_df['Total Purchase Value'] = purchasing_age_df['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_age_df['Avg Total Purchase per Person'] = purchasing_age_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)


purchasing_age_df





```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_group_df = purchase_data.groupby("SN")

purchase_count = spender_group_df["SN"].count()

average_purch_price = round((spender_group_df["Price"].sum()/spender_group_df["SN"].count()), 2)
average_purch_price

total_purch_value = spender_group_df["Price"].sum()

top_spender_df = pd.DataFrame({"Purchase Count": purchase_count,
                               "Average Purchase Price": average_purch_price,
                               "Total Purchase Value": total_purch_value})


top_spender_df['Average Purchase Price'] = top_spender_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spender_df['Total Purchase Value'] = top_spender_df['Total Purchase Value'].map('${:,.2f}'.format)

top_spender_sorted = top_spender_df.sort_values(by=["Total Purchase Value"], ascending = False)
top_spender_sorted.head()


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data[['Item ID', 'Item Name', 'Price']]
popular_group_df = popular_df.groupby(["Item ID","Item Name"]).agg(['count', 'sum', 'mean'])
renamed_popular_df = popular_group_df.rename(columns={"count": "Purchase Count", "mean": "Item Price", "sum": "Total Purchase Value" })


renamed_popular_df.head(10)

#### QUESTION FOR TA - not able to put $ sign as above using format
                                 
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python

renamed_popular_df.columns = renamed_popular_df.columns.map("_".join)

sort_popular_df = renamed_popular_df.sort_values(by=["Price_Total Purchase Value"], ascending = False)

sort_popular_df['Price_Total Purchase Value'] = sort_popular_df['Price_Total Purchase Value'].map('${:,.2f}'.format)
sort_popular_df['Price_Item Price'] = sort_popular_df['Price_Item Price'].map('${:,.2f}'.format)

sort_popular_df.head(10)
```
